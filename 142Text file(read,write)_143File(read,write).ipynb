```
#Unite 142 예제
f=open('stockcode.txt','r')
h=open('stockcode_copy.txt','w')

data=f.read()
h.write(data)

f.close()
h.close()
```

```
#Unite 143 예제
bufsize=1024
f=open('img_sample.jpg','rb')
h=open('img_sample_copy.jpg','wb')

data=f.read(bufsize)
while data:
  h.write(data)
  data=f.read(bufsize)

f.close()
h.close()
```
