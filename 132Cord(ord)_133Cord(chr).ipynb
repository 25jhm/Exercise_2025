```
#Unite 132 예제
ch=input('문자 1개를 입력하세요:')
if len(ch)!=0:
  ch=ch[0]
  chv=ord(ch)
  print('문자%s\t코드값:%d[%s]'%(ch,chv,hex(chv)))
```

```
#Unite 133 예제
va1=input('문자 코드값을 입력하세요:')
va1=int(va1)
try:
  ch=chr(va1)
  print('코드값:%d[%s],문자:%s'%(va1,hex(va1),ch))
except ValueError:
  print('입력한 <%d>에 대한 문자가 존재하지 않습니다!'%va1)
```
