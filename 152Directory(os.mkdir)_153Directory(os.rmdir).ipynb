```
#Unite 152 예제
import os

newfolder=input('새로운 생성할 디렉터리 이름을 입력핳세요:')
try:
  os.mkdir(newfolder)
  print('[%s]디렉터리를 새로 생성했습니다.')
except Exception as e:
  print(e)
```

```
#Unite 153 예제
import os

target_folder='tmp'
k=input('[%s]디렉터리를 삭제하시겠습니까?(y/n)'%target_folder)
if k=='y':
  try:
    os.rmdir(target_folder)
    print('[%s]디렉터리를 삭제했습니다.'%target_folder)
  except Exception as e:
   print(e)
```
