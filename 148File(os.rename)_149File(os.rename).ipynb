```
#Unite 148 예제
from os import rename

target_file='stokcode.txt'
newname=input('[%s]에 대한 새로운 파일이름을 입력하세요:'%target_file)
rename(target_file,newname)
print('[%s]->[%s]로 파일이 변경되었습니다.'%(target_file,newname))
```

```
#Unite 149 예제
from os import rename

target_file='stockcode.txt'
newpath=input('[%s]를 이용할 디렉터리의 절대경로를 입력하세요:'%target_file)

if newpath[-1]=='/':
  newname=newpath+target_file
else:
  newname=newpath+'/'+target_file

try:
  rename(target_file,newname)
  print('[%s]->[%s]로 이동되었습니다.'%(target_file,newname))
except FileNotFoundError as e:
  print(e)
```
