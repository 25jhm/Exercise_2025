```
#Unite 123 예제
solar1=['태양','수성','금성','지구','화성','목성','토성','청왕성','해왕성']
solar2=['Sun','Meecury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
solardict={}
for i,k in enumerate(solar1):
  va1=solar2[i]
  solardict[k]=va1

print(solardict)
```
