```
#Unite 164 예제
url='http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sidl=105&oid=028&aid=0002334601'

tmp=url.split('?')
queries=tmp[1].split('&')
for query in queries:
  print(query)
```

```
#Unite 165 예제
mystack=[]

def putdata(data):
  global mystack
  mystack.append(data)

def popdata():
   global mystack
   if len(mystack)==0:
    return None
   return mystack.pop()

putdata('데이터1')
putdata([3,4,5,6])
putdata(12345)

print('<스택상태>:',end='');print(mystack)

ret=popdata()
while ret !=None:
  print('스택에서 데이터 추출:',end='');print(ret)
  print('<스택상태>:',end='');print(mystack)
  ret=popdata()
```
